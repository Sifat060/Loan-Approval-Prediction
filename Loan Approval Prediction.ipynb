```python

import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report
from sklearn.metrics import accuracy_score
from sklearn.model_selection import train_test_split
```

```python
# Dropping = ['Loan_ID']
# volunteer = volunteer.drop(Dropping,axis=1)
X = volunteer.iloc[:, :-2]
y = volunteer.iloc[:, -2]
```

```python
x_train, x_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=20)
```

```python
#Train the model
model = LogisticRegression()
model.fit(x_train, y_train) #Training the model
predictions = model.predict(x_test)
print(predictions)# printing predictions

# Dropping = ['Loan_ID']
# x_train = x_train.drop(Dropping,axis=1)
#X
```

```python
print("Training accuracy of the model is {:.2f}".format(model.score(x_train, y_train)))
print("Testing accuracy of the model is {:.2f}".format(model.score(x_test, y_test)))
```

**Decision Tree**



```python
from sklearn.tree import DecisionTreeClassifier
DTC = DecisionTreeClassifier(criterion='entropy',random_state=0)
DTC.fit(x_train, y_train)
DTC_predict = DTC.predict(x_test)
print(DTC_predict)
```

```python
from sklearn import tree
print("Training accuracy of the model is {:.2f}".format(DTC.score(x_train, y_train)))
print("Testing accuracy of the model is {:.2f}".format(DTC.score(x_test, y_test)))
fig, axes = plt.subplots(nrows = 1,ncols = 1,figsize = (4,4), dpi=300)
tree.plot_tree(DTC,
               feature_names = X.columns, 
               class_names=['1','2','3','4','5','6','7'],
               filled = True);
```
